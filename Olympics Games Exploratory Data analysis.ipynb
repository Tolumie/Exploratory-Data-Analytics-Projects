```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load dataset
athletes = pd.read_csv('athlete_events.csv')
region = pd.read_csv('noc_regions.csv')
athletes.head()
```

```python
region.head()
```

```python
# join the dataset
athletes_df = athletes.merge(region, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
#shape to know nos of rows and colmns
athletes_df.head()
```

```python
# make column names consistent
athletes_df.rename(columns = {'region': 'Region', 'notes' : 'Notes'}, inplace = True);
```

```python
# Top ten countries participating since Olympics inception since 1986

top_10_countries =  athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
print(top_10_countries)
```

```python
#plot for the top ten countries

plt.figure(figsize= (12,6))
#plt.xtricks(rotation = 20)
plt.title('Overall participation by country')
sns.barplot(x = top_10_countries.index, y = top_10_countries, palette = "Set2")
```

```python
# Age distribution of the participants

plt.figure(figsize= (12,6))
plt.title('Age distribution of the participants')
plt.xlabel ('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange',  edgecolor = 'white');
```

```python
# Gold medal athletes
Gold_Medals = athletes_df[(athletes_df.Medal == 'Gold')]
Gold_Medals.head()
```

```python
# take only the values that are different from NaN: values only

Gold_Medals = Gold_Medals[np.isfinite(Gold_Medals['Age'])]
```

```python
# GOld medalists with age beyond 60: count
Gold_Medals['ID'][Gold_Medals['Age'] > 60].count()
```

```python
# GOld medalists with age beyond 60: sporting_event
sporting_event = Gold_Medals[['Sport']][Gold_Medals['Age'] > 60]
sporting_event
```

```python
# GOld medalists with age beyond 60: sporting_event with few deails
sporting_event = Gold_Medals[['Sport','Name','Region','Year']][Gold_Medals['Age'] > 60]
sporting_event
```

```python
# plot for GOld medalists with age beyond 60
plt.figure(figsize =(10,5))
plt.tight_layout()
sns.countplot(x = sporting_event.Sport )
plt.title('Gold Medals for Athletes over 60 years')
plt.show()
```

```python
# Gold medals for each country
Gold_Medals = athletes_df[(athletes_df.Medal == 'Gold')]
```

```python
Gold_Medals.Region.value_counts().reset_index(name = 'Medal').head()
```

```python
totalgoldmedals = Gold_Medals.Region.value_counts().reset_index(name = 'Medal').head(7)
g = sns.catplot(x = 'index', y ='Medal', data = totalgoldmedals,
               height = 5, kind = 'bar', palette = 'rocket')
g.despine(left = True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per Country')
```
